```python
import os
import pandas as pd
```

```python
cd = os.getcwd()
```

```python
def remove_words(category, words_to_remove):
    for word in words_to_remove:
        category = category.replace(word, '')
    category = category.strip()
    return category
```

```python
files = []
topics, speakers, stimes, etimes, locs = [], [], [], [], []
words_to_remove = ['<sentence>', '</sentence>', '.', '<paragraph>', '</paragraph>']
for file in os.listdir(cd + "\\sa-tagged"):
    files.append(file)
    tmp_file = open(cd + "\\sa-tagged\\" + file, 'r')
    tmp_lines = tmp_file.readlines()
    (c1, c2, c3, c4, c5) = (0, 0, 0, 0, 0)
    (topic, speaker, stime, etime, location) = ('', '', '', '', '')
    for line in tmp_lines:
        if (('Topic:' in line) & (c1 != 1)):
            topic = line.split('Topic:')[1]
            topic = topic.strip()
            topic = topic.replace('\n', '')
            topics.append(topic)
            c1 += 1
        if (('<speaker>' in line) & (c2 != 1)):
            speaker = line.split('<speaker>')[1]
            speaker = speaker.split('</speaker')[0]
            speaker = remove_words(speaker, words_to_remove)
            speakers.append(speaker)
            c2 += 1
        if (('<stime>' in line) & (c3 != 1)):
            stime = line.split('<stime>')[1]
            stime = stime.split('</stime>')[0]
            stime = remove_words(stime, words_to_remove)
            stimes.append(stime)
            c3 += 1
        if (('<etime>' in line) & (c4 != 1)):
            etime = line.split('<etime>')[1]
            etime = etime.split('</etime>')[0]
            etime = remove_words(etime, words_to_remove)
            etimes.append(etime)
            c4 += 1
        if (('<location>' in line) & (c5 != 1)):
            location = line.split('<location>')[1]
            location = location.split('</location>')[0]
            location = remove_words(location, words_to_remove)
            locs.append(location)
            c5 += 1
    list_of_ctgs = [topics, speakers, stimes, etimes, locs]
    list_of_ctgs2 = [topic, speaker, stime, etime, location]
    for i in range(len(list_of_ctgs2)):
        if list_of_ctgs2[i] == '':
            list_of_ctgs[i].append(list_of_ctgs2[i])
```

```python
seminar_df = pd.DataFrame({'Speaker': speakers,
                           'Topic': topics,
                           'Start time': stimes,
                           'End time': etimes,
                           'Location': locs}, index=files)
seminar_df.index.name = 'Name of file'
seminar_df
```

```python
seminar_df.to_csv("seminar_announcements.csv",
                  sep=';')
```
